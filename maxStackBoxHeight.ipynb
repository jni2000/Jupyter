```python
# Dynamic Programming implementation 
# of Box Stacking problem: calculate the max possible stack height of boxes stacked up with 
# strict size comparisoion (i.e. the w, h, d of an upper box must be smaller than the w ,h, d of a lower box)
# rotate is allowed or not allowed

class Box: 
      
    # Representation of a box 
    def __init__(self, w, h, d): 
        self.h = h 
        self.w = w 
        self.d = d 
    # less than
    def __lt__(self, other): 
        return (self.d < other.d and self.w < other.w and self.h < other.h) 
  
def maxStackHeight(arr, n, rotate): 
  
    # Create an array of all rotations of  
    # given boxes. For example, for a box {1, 2, 3},  
    # we consider three instances{{1, 2, 3}, 
    # {2, 1, 3}, {3, 1, 2}} 
    rot = [Box(0, 0, 0) for _ in range(3 * n)] 
    index = 0
  
    for i in range(n): 
        if rotate:
            # Copy the original box 
            rot[index].h = arr[i].h 
            rot[index].d = max(arr[i].d, arr[i].w) 
            rot[index].w = min(arr[i].d, arr[i].w) 
            index += 1

            # First rotation of the box 
            rot[index].h = arr[i].w 
            rot[index].d = max(arr[i].h, arr[i].d) 
            rot[index].w = min(arr[i].h, arr[i].d) 
            index += 1

            # Second rotation of the box 
            rot[index].h = arr[i].d 
            rot[index].d = max(arr[i].h, arr[i].w) 
            rot[index].w = min(arr[i].h, arr[i].w) 
            index += 1
        else:
            rot[index] = arr[i]
            index += 1
            rot[index] = arr[i]
            index += 1
            rot[index] = arr[i]
            index += 1
            
        
    # Now the number of boxes is 3n 
    n *= 3
  
    # Sort the array 'rot[]' in non-increasing  
    # order of base area 
    rot.sort(reverse = True) 
  
    # Uncomment following two lines to print  
    # all rotations  
    # for i in range(n): 
    #     print(rot[i].h, 'x', rot[i].w, 'x', rot[i].d) 
  
    # Initialize msh values for all indexes 
    # msh[i] --> Maximum possible Stack Height  
    # with box i on top 
    msh = [0] * n 
  
    for i in range(n): 
        msh[i] = rot[i].h 
  
    # Compute optimized msh values 
    # in bottom up manner 
    for i in range(1, n): 
        for j in range(0, i): 
            if (rot[i].w < rot[j].w and 
                rot[i].d < rot[j].d and
                rot[i].h < rot[j].h): 
                if msh[i] < msh[j] + rot[i].h: 
                    msh[i] = msh[j] + rot[i].h 
  
    maxm = -1
    for i in range(n): 
        maxm = max(maxm, msh[i]) 
  
    return maxm 
  
# Driver Code 
if __name__ == "__main__": 
    arr = [Box(2, 1, 7), Box(1, 2, 3), 
           Box(4, 5, 6), Box(5, 8, 9)] 
    n = len(arr) 
    print("The maximum possible height of stack is", 
           maxStackHeight(arr, n, False))
```
