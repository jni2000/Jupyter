```python
"""
A shopkeeper has a sale to complete and has arranged the items being sold in a list. Starting from the left, the shop 
keeper rings up each item at its full price less the price of the first lower or equally priced item to its right. 
If there is no item to the right that costs less than or equal to the current item's price the current item is sold at 
full price.

For example, assume there are items priced [2, 3, 1, 2, 4, 2].
The first and second items would each be discounted by 1 unit, the first equal or lower price to the right.
The item priced 1 unit would sell at a full price.
The next item, at 2 units, would be discounted 2 units as would the 4 unit item.
The sixth and final item must be purchased at full price.
The total cost is 1+2+1+0+2+2 = 8 units.

Print total cost of all items on the first line.
On the second line print a space separated list of integers representing the indexes of the non- discounted items in 
ascending index order.

1 <= size(prices) <= 100000
1 <= prices <= 1000000
Input: [2, 3, 1, 2, 4, 2]
Output:
8
2 5

Input 2: [5,1,3,4,6,2]
Output:
14
1 5

Input 3: [1,3,3,2,5]
Output:
9
0 3 4
"""


# returns the final prices of all items after discounts
def finalPrice(prices):
    if len(prices) == 0:
        return 0

    nle_stack = [] # nle = next lesser element
    res = 0
    saleprices = []
    # we go from right to left
    for i in reversed(range(len(prices))):
        # pop the stack until we find the next lesser element
        while nle_stack and nle_stack[-1][0] > prices[i]:
            nle_stack.pop()

        # add the discount, if any, and then add the current price to the stack
        marked_price = prices[i] if not nle_stack else prices[i] - nle_stack[-1][0]
        res += marked_price
        saleprices.append(marked_price)
        nle_stack.append([prices[i], i])
    finalprice = saleprices[::-1]
    unchanged = [prices[i] ^ finalprice[i] for i in range(len(prices))]
    #print(prices)
    #print(finalprice)
    #print((unchanged))
    original_priced = ""
    for i in range(len(prices)):
        if unchanged[i] == 0:
            original_priced += (str(i) + " ")
    print(original_priced)
    return res

def finalPrice2(prices):
    stack = []
    res = 0
    for i in prices:
        if(stack):
            if i > stack[-1]:
                stack.append(i)
            else:
                while(stack and i <= stack[-1]):
                    res += stack[-1] - i
                    stack.pop(-1)
                stack.append(i)
        else:
            stack.append(i)
    return res + sum(stack)
pricelist1 = [2, 3, 1, 2, 4, 2]
pricelist2 = [5,1,3,4,6,2]
pricelist3 = [1,3,3,2,5]

print("1:",finalPrice(pricelist1))
print("1:",finalPrice(pricelist2))
print("1:",finalPrice(pricelist3))
print("2:", finalPrice2(pricelist))
```
