```python
import folium
import os
merc = os.path.join('C:/Users/James/Sam-1.png')
m = folium.Map([40.00, -74.21], zoom_start=1, tiles='stamentoner')
if not os.path.isfile(merc):
    print(f'Could not find {merc}')
else:
    img = folium.raster_layers.ImageOverlay(
        name='Mercator projection SW',
        image=merc,
        bounds=[[-82, -180], [82, 180]],
        opacity=0.6,
        interactive=True,
        cross_origin=False,
        zindex=1,
    )

    folium.Popup('I am an image').add_to(img)

    img.add_to(m)
    folium.LayerControl().add_to(m)
m
```
