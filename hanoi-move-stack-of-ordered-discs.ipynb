```python
"""
Recursive Python function to solve tower of hanoi
"""
class Disc:
    def __init__(self, dimension):
        self.dimension=dimension
    def dimension(self):
        return self.dimension
    def show(self):
        print("Disc:", self.dimension) 
        
class Rod:
    def __init__(self, label):
        self.label=label
        self.rod = []
    def push(self, disc):
        if disc:
            self.rod.append(disc)
    def pop(self):
        disc = None
        if len(self.rod) > 0:
            disc = self.rod.pop()
        return disc
    
    def show(self):
        print("Rod: ", self.label)
        for i in range(len(self.rod)):
            print(i, self.rod[i].dimension)

A = Rod("A")
B = Rod("B")
C = Rod("C")
A.push(Disc(4))
A.push(Disc(3))
A.push(Disc(2))
A.push(Disc(1))

def TowerOfHanoi(n , from_rod, to_rod, aux_rod):
    if n == 1: 
        print ("Move disk 1 from rod",from_rod.label,"to rod",to_rod.label)
        disc=from_rod.pop()
        to_rod.push(disc)
        return
    TowerOfHanoi(n-1, from_rod, aux_rod, to_rod) 
    print ("Move disk",n,"from rod",from_rod.label,"to rod",to_rod.label)
    disc=from_rod.pop()
    to_rod.push(disc)
    TowerOfHanoi(n-1, aux_rod, to_rod, from_rod) 
    
n=4
A.show()
B.show()
C.show()
TowerOfHanoi(n, A, C, B)
A.show()
B.show()
C.show()
```
