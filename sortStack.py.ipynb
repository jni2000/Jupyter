```python
"""
sort stack using two stacks
"""

stack1 = []
stack2 = []

stack1.append(3)
stack1.append(2)
stack1.append(1)
stack1.append(0)
stack1.append(4)
stack1.append(5)

last_element=-1
done = False
while not done:
    for i in range(len(stack1)):
        print(stack1)
        print(stack2)
        element = stack1.pop()
        if i == 0:
            stack2.append(element)
            last_element = element
        else:
            if element < last_element:
                if len(stack1) !=0:
                    next_element = stack1.pop()
                    stack1.append(stack2.pop())
                    stack2.append(element)
                    stack1.append(next_element)
                else:
                    next_element = stack2.pop()
                    stack2.append(element)
                    stack2.append(next_element)
                    done = True
                last_element=element
            else:
                if len(stack1) !=0:
                    next_element = stack1.pop()
                    stack1.append(element)
                    stack1.append(next_element)
                else:
                    next_element = None                    
                    stack2.append(element)                    
                    done = True
print(stack1)
print(stack2)
```
