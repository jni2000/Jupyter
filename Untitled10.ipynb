```python
"""
find the maximum path score of the paths from upper left conner to the lower right
conner of the matrix.

The score of of a path the is minimum value along the path
"""
matrix_data = [[7,5,3],
               [2,0,9],
               [4,5,9]]
rightDone = [False]*(len(matrix_data)*len(matrix_data[0]))
downDone = [False]*(len(matrix_data)*len(matrix_data[0]))
rightDone[(len(matrix_data)*len(matrix_data[0])-1)] = True
downDone[(len(matrix_data)*len(matrix_data[0])-1)] = True

def start(matrix, i,j):
    return i==0 and j==0

def end(matrix, i,j):
    return i==len(matrix)-1 and j==len(matrix[0])-1

def label(matrix, i,j):
    if i<=len(matrix)-1 and j<=len(matrix[0])-1:
        return i*len(matrix[0]) + j
    else:
        return None
    
def rightMoveable(matrix, i,j):
    move = True
    if j>=len(matrix[0])-1:
        move=False
    if i>=len(matrix):
        move = False
    return move

def downMoveable(matrix, i,j):
    move = True
    if i>=len(matrix)-1:
        move=False
    if j>=len(matrix[0]):
        move = False
    return move

def moveRight(matrix, i, j):
    if rightMoveable(matrix, i,j):
        return [i, j+1]
    else:
        return None
    
def moveDown(matrix, i, j):
    if downMoveable(matrix, i,j):
        return [i+1, j]
    else:
        return None

def rightVisited(matrix, i, j):
    index = label(matrix, i,j)
    if index is not None:
        return rightDone[index]
    else:
        return True
    
def downVisited(matrix, i, j):
    index = label(matrix, i,j)
    if index is not None:
        return downDone[index]
    else:
        return True

def fullyVisited(matrix, i, j):
    visited = ((rightMoveable(matrix, i,j) and rightVisited(matrix, i, j)) or (not rightMoveable(matrix, i,j))) and ((downMoveable(matrix, i,j) and downVisited(matrix, i, j)) or (not downMoveable(matrix, i,j)))
        
    return visited

def visitRight(matrix, i, j):
    index = label(matrix, i,j)
    if index is not None:
        rightDone[index] = True
    
def visitDown(matrix, i, j):
    index = label(matrix, i,j)
    if index is not None:
        downDone[index] = True

def unvisitRight(matrix, i, j):
    index = label(matrix, i,j)
    if index is not None:
        rightDone[index] = False
    
def unvisitDown(matrix, i, j):
    index = label(matrix, i,j)
    if index is not None:
        downDone[index] = False
        
def move(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if start(matrix, i,j):
                print('start:', [i,j])
            if end(matrix, i,j):
                print('end:', [i,j])
            moves = [rightMoveable(matrix, i,j), downMoveable(matrix, i,j)]
            print(label(matrix, i,j), moves)
            
def findPaths(matrix):
    paths =[]
    stack=[]
    i=0
    j=0
    done=False
    print(rightDone)
    print(downDone)
    
    stack.append([i,j])
    while not done:
        print("go:", i, j, stack)
        if rightMoveable(matrix, i,j) and not rightVisited(matrix, i, j):
            visitRight(matrix, i, j)
            nextNode = moveRight(matrix, i, j)
            if nextNode is not None:
                i=nextNode[0]
                j=nextNode[1]
                stack.append([i,j])
        elif downMoveable(matrix, i,j) and not downVisited(matrix, i, j):
            visitDown(matrix, i, j)
            nextNode = moveDown(matrix, i, j)
            if nextNode is not None:
                i=nextNode[0]
                j=nextNode[1]
                stack.append([i,j])
        else:
            if len(stack) != 0:
                nextNode = stack.pop()
                if nextNode is not None:
                    i=nextNode[0]
                    j=nextNode[1]
                    if not fullyVisited(matrix, i, j):
                        stack.append([i,j])
                    else:
                        unvisitRight(matrix, i, j)
                        unvisitDown(matrix, i, j)
            else:
                done = True
    return paths

if __name__ == '__main__':
    print(findPaths(matrix_data))
    

```
