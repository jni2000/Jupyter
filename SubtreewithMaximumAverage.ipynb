```python
"""
Subtree with Maximum Average
Given a M-ary tree, find the subtree with maximum average. Return the root of the subtree.
A subtree of a tree is any node of that tree plus all its descendants. The average value of 
a subtree is the sum of its values, divided by the number of nodes.

Input:
		 20
	   /   \
	 12     18
  /  |  \   / \
11   2   3 15  8

Output: 18

Input:
         1
        /|\
      -5 13 4
      /\ /\
     1 2 4 -2
Output: 13    
"""

"""Solutiuon 1"""
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.children = []
        
    def addChild(self, c):
        if c:
            self.children.append(c)

class Solution: 
    def MaxAverageSubtree(self, root):
        if not root or not root.children:
            return None
        
        self.res = [float('-inf'), 0]
        # self.res[0]: average; self.res[1]: number of nodes
        self.dfs(root)
        return self.res[1]
    
    def dfs(self, root):
        if not root.children:
            return [root.val, 1]
        
        temp_sum, temp_num = root.val, 1
        for child in root.children:
            child_sum, child_num = self.dfs(child)
            temp_sum += child_sum
            temp_num += child_num
            
        if temp_sum/temp_num > self.res[0]:
            self.res = [temp_sum/temp_num, root.val]
        
        return [temp_sum, temp_num]

"""
n1 = TreeNode(8)
n2 = TreeNode(15)
n3 = TreeNode(3)
n4 = TreeNode(2)
n5 = TreeNode(11)
n6 = TreeNode(18)
n6.addChild(n1)
n6.addChild(n2)
n7 = TreeNode(12)
n7.addChild(n3)
n7.addChild(n4)
n7.addChild(n5)
n8 = TreeNode(20)
n8.addChild(n6)
n8.addChild(n7)

root = n8

s = Solution()
s.MaxAverageSubtree(n8)


n11 = TreeNode(1)
n12 = TreeNode(2)
n13 = TreeNode(4)
n14 = TreeNode(-2)
n15 = TreeNode(-5)
n15.addChild(n11)
n15.addChild(n12)
n16 = TreeNode(13)
n16.addChild(n13)
n16.addChild(n14)
n17 = TreeNode(4)
n18 = TreeNode(1)
n18.addChild(n15)
n18.addChild(n16)
n18.addChild(n17)

root = n18
s = Solution()
s.MaxAverageSubtree(n18)
"""


"""Solution 2"""
from collections import deque
class TreeNode2:
    def __init__(self, val, children):
        self.val = val
        self.children = children
        self.total_val = val
        self.total_num = 1

class Solution2:
    def findmaxavgsubtree2(self, root: 'TreeNode') -> 'TreeNode':

        stack1 = deque([])
        stack2 = deque([])
        stack1.append(root)

        max_avg = float('-inf')
        max_avg_node = None

        while stack1:
            visit = stack1.pop()
            stack2.append(visit)
            for child in visit.children:
                stack1.append(child)

        while stack2:
            visit = stack2.pop()
            # if not leaf node, calculate total value and total number
            if len(visit.children) > 0:
                for child in visit.children:
                    visit.total_val += child.total_val
                    visit.total_num += child.total_num
                cur_avg = visit.total_val/visit.total_num
                if cur_avg > max_avg:
                    max_avg = cur_avg
                    max_avg_node = visit
        return max_avg_node

if __name__ == '__main__':

    n4 = TreeNode2(11, [])
    n5 = TreeNode2(2, [])
    n6 = TreeNode2(3, [])
    n7 = TreeNode2(15, [])
    n8 = TreeNode2(8, [])
    n2 = TreeNode2(12, [n4, n5, n6])
    n3 = TreeNode2(18, [n7, n8])
    n1 = TreeNode2(20, [n2, n3])

    ss = Solution2()
    print(ss.findmaxavgsubtree2(n1).val)

    n11 = TreeNode2(1,[])
    n12 = TreeNode2(2,[])
    n13 = TreeNode2(4,[])
    n14 = TreeNode2(-2,[])
    n15 = TreeNode2(-5, [n11,n12])
    n16 = TreeNode2(13, [n13,n14])
    n17 = TreeNode2(4,[])
    n18 = TreeNode2(1,[n16,n16,n17])

    ss = Solution2()
    print(ss.findmaxavgsubtree2(n18).val)

```
