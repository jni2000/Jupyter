```python
"""
find the maximum path score of the paths from upper left conner to the lower right
conner of the matrix.

The score of of a path the is minimum value along the path

Method: using recursive from 1x1 array.
the path score and max path score of an 1x1 array is 0, S(1)
the path max path score of a 2x2 array shall be the max(S(1), score(added edge path))
"""
matrix_data = [[6,5,3],
               [2,0,9],
               [4,7,9],
               [3,1,5]]

class Node:
    def __init__(self, key, value, neighbors, visited):
        self.key = key
        self.value = value 
        self.neighbors=neighbors
        self.visited = 0
        
def buildNodes(matrix, nodes):
    for i in reversed(range(len(matrix))):
        for j in reversed(range(len(matrix[0]))):
            neighbors = []
            if i<len(matrix)-1:
                neighbors.append(nodes[i+1][j])
            if j<len(matrix[0])-1:
                neighbors.append(nodes[i][j+1])
            nodes[i][j] = Node((i*len(matrix[0]))+j, matrix[i][j], neighbors, 0)            
            
def endNode(node, end):
    if node.key == end.key:
        return True
    return False

def printNode(node):
    print("key:", node.key)
    print("value:", node.value)
    print("neighbors:", len(node.neighbors))
    for i in range(len(node.neighbors)):
        print("v:", node.neighbors[i].value)
    print("visited:", node.visited)
    print("..........")
    
def printNodes(nodes):
    for i in range(len(nodes)):
        for j in range(len(nodes[0])):
            printNode(nodes[i][j])
        print("==========")
            
def printStack(mark, stack):
    for i in range(len(stack)):
        print(mark, stack[i].key)
        
def findPath(node, end):
    if node.visited < len(node.neighbors) or endNode(node, end):
        node.visited += 1
        path.append(node.value)
        stack.append(node)
        if endNode(node, end):
            paths.append([path.copy(), min(path)])
            print(path)
            done = False
            while not done:
                if len(stack) == 0:
                    done = True
                else:
                    n = stack.pop()
                    if len(n.neighbors) > n.visited and end not in n.neighbors and not endNode(n, end):
                        n.visited +=1
                        stack.append(n)
                        done = True
                    else:
                        n.visited = False
                        path.pop()
    for n in node.neighbors:
        findPath(n, end)    

if __name__=="__main__":
    nodes = [[Node(0, 0, None, 0) for j in range(len(matrix_data[0]))] for i in range(len(matrix_data))]
    buildNodes(matrix_data, nodes)
    printNodes(nodes)
    
    paths=[]
    path=[]
    stack=[]
    findPath(nodes[0][0], nodes[len(matrix_data)-1][len(matrix_data[0])-1])
    print(paths)
    print("max path score is: ", max([paths[i][1] for i in range(len(paths))]))

    
```
